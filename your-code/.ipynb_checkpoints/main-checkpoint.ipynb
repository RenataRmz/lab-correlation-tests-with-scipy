# Challenge 1 - Analysis of Variance

In this part of the lesson, we will perform an analysis of variance to determine whether the factors in our model create a significant difference in the group means. We will be examining a dataset of FIFA players. We'll start by loading the data using the code in the cell below.

```python
# Run this code:
fifa = pd.read_csv('fifa.csv')
```

Let's examine the dataset by looking at the `head`.

```python
# Your code here:
fifa.head()
```

Player's values are expressed in millions of euros. We would like this column to be numeric. Therefore, let's create a numeric value column. Do this by stripping all non-numeric characters from each cell. Assign this new data to `ValueNumeric`. There is no need to multiply the value to be expressed in millions. 

```python
#Quita el símbolo de euros, millones(M) y miles(K)
fifa['ValueNumeric_']= fifa['Value'].str.replace('€','').str.replace('M','').str.replace('K','').astype('float64')
#Identifica si el monto está en millones(M) o miles(K)
fifa['Decimales'] = fifa['Value'].str[-1]
#Si el monto es en miles, entonces divide entre mil para llevarlo a millones
fifa['ValueNumeric'] = np.where(fifa['Decimales']=='K',fifa['ValueNumeric_']/1000,fifa['ValueNumeric_'])
#Valida los cambios
fifa
```

#### We'd like to determine whether a player's preffered foot and position have an impact on their value. 

Using the `statsmodels` library, we are able to produce an ANOVA table without munging our data. Create an ANOVA table with value as a function of position and preferred foot. Recall that pivoting is performed by the `C` function.

Hint: For columns that have a space in their name, it is best to refer to the column using the dataframe (For example: for column `A`, we will use `df['A']`).

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
#Definición de hipótesis
#H_0: mu_1 = mu_2 = ... = mu_3
#H_1: At least one mean is different
#Renombra Preferred Foot para que no tenga espacios
fifa = fifa.rename(columns={"Preferred Foot":"Preferred_foot"})
#Define la fórmula
formula = "ValueNumeric ~ C(Position)+C(Preferred_foot)"
#Crea el modelo
model = ols(formula,data=fifa).fit()
anova_table = sm.stats.anova_lm(model,typ=2)
anova_table
```

What is your conclusion from this ANOVA?

```python
# Your conclusions here:
#Dado que p-value es menor a 0.05 en ambos casos, se rechaza la hipótesis nula 
#y concluimos que el valor de los jugadores cambia de acuerdo a la posición y 
#el pie de preferencia con el que juegan.
```

After looking at a model of both preffered foot and position, we decide to create an ANOVA table for nationality. Create an ANOVA table for numeric value as a function of nationality.

```python
#Definición de hipótesis
#H_0: mu_1 = mu_2 = ... = mu_3
#H_1: At least one mean is different
#Define la fórmula
formula = "ValueNumeric ~ C(Nationality)"
#Crea el modelo
model = ols(formula,data=fifa).fit()
anova_table = sm.stats.anova_lm(model,typ=2)
anova_table
```

What is your conclusion from this ANOVA?

```python
#Dado que p-value es menor a 0.05, se rechaza la hipótesis nula 
#y concluimos que el valor de los jugadores cambia de acuerdo a su nacionalidad.
```

# Challenge 2 - Linear Regression

Our goal with using linear regression is to create a mathematical model that will enable us to predict the outcome of one variable using one or more additional independent variables.

We'll start by ensuring there are no missing values. Examine all variables for all missing values. If there are missing values in a row, remove the entire row.

```python
# Your code here:
fifa.isnull().sum()
```

```python
fifa = fifa.dropna()
```

```python
#Valida
fifa.isnull().sum()
```

Using the FIFA dataset, in the cell below, create a linear model predicting value using stamina and sprint speed. create the model using `statsmodels`. Print the model summary.

Hint: remember to add an intercept to the model using the `add_constant` function.

```python
import statsmodels.api as sm
X = sm.add_constant(fifa[['SprintSpeed', 'Stamina']])
Y = fifa.ValueNumeric

model = sm.OLS(Y,X).fit()
predictions = model.predict(X)

print_model = model.summary()
print(print_model)
```

Report your findings from the model summary. In particular, report about the model as a whole using the F-test and how much variation is predicted by the model using the r squared.

```python
# Your conclusions here:
#La ecuación de regresión es la siguiente:
#ValueNumeric = -3.0700 + 0.0263 * SprintSpeed + 0.604 * Stamina

#La r-cuadrada es 0.048, la cual es pequeña. Esto significa que el modelo únicamente 
#captura el 4.8% de las variaciones de los datos.

#El p-value es muy pequeño 8.04e-193, esto significa que la pendiente es 
#significativamente distinta de cero.
```

Next, create a second regression model predicting value using potential. Create the model using `statsmodels` and print the model summary. Remember to add a constant term.

```python
# Your code here:
X = sm.add_constant(fifa['Potential'])
Y = fifa.ValueNumeric

model = sm.OLS(Y,X).fit()
predictions = model.predict(X)

print_model = model.summary()
print(print_model)
```

Report your findings from the model summary. In particular, report about the model as a whole using the F-test and how much variation is predicted by the model using the r squared.

```python
# Your conclusions here:
#La ecuación de regresión es la siguiente:
#ValueNumeric = -35.4390 + 0.5312 * Potential

#La r-cuadrada es 0.336. Esto significa que el modelo  
#captura el 33.6% de las variaciones de los datos.

#El p-value es cero, esto significa que la pendiente es 
#significativamente distinta de cero.
```
